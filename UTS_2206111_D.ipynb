# HISTOGRAM CITRA


````python
# Import library yang diperlukan
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menampilkan gambar
def show_image(image, title, cmap='gray'):
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Unggah citra untuk analisis
from google.colab import files
uploaded = files.upload()

# Membaca citra yang diunggah (gunakan file pertama yang diunggah)
file_name = list(uploaded.keys())[0]
image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

# Menampilkan citra grayscale
show_image(image, "Citra Grayscale")

# Perhitungan histogram
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Normalisasi histogram
hist_normalized = hist / hist.sum()

# Visualisasi histogram asli
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(hist, color='blue')
plt.title('Histogram Asli')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi')

# Visualisasi histogram normalisasi
plt.subplot(1, 2, 2)
plt.plot(hist_normalized, color='green')
plt.title('Histogram Ternormalisasi')
plt.xlabel('Intensitas')
plt.ylabel('Frekuensi Relatif')

plt.tight_layout()
plt.show()
````

# TRANSFORMASI FOURIER

````python
# Import library yang diperlukan
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Unggah gambar dari komputer
from google.colab import files
uploaded = files.upload()

# Membaca gambar grayscale
image_path = list(uploaded.keys())[0]  # Nama file pertama yang diunggah
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Menampilkan gambar asli
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Melakukan Transformasi Fourier
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20 * np.log(np.abs(fshift))

# Menampilkan magnitude spectrum
plt.subplot(1, 3, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Magnitude Spectrum")
plt.axis('off')

# Melakukan invers Transformasi Fourier
f_ishift = np.fft.ifftshift(fshift)
img_reconstructed = np.fft.ifft2(f_ishift)
img_reconstructed = np.abs(img_reconstructed)

# Menampilkan gambar hasil invers transformasi
plt.subplot(1, 3, 3)
plt.imshow(img_reconstructed, cmap='gray')
plt.title("Reconstructed Image")
plt.axis('off')

plt.tight_layout()
plt.show()
````

### **5. Transformasi Fourier**

  - Hasilnya berupa nilai kompleks yang berisi informasi magnitude (amplitudo) dan fase (sudut).
- `np.fft.fftshift(f)`: Menggeser frekuensi nol (frekuensi rendah) ke tengah gambar untuk visualisasi yang lebih mudah.
- `20 * np.log(np.abs(fshift))`: Menghitung **magnitude spectrum** dalam skala logaritmik agar mudah divisualisasikan. Nilai magnitude diperoleh dari modulus nilai kompleks.

- `np.abs(img_reconstructed)`: Mengambil nilai real dari hasil invers, mengabaikan kemungkinan adanya nilai imajiner kecil akibat error numerik.

# OPERASI ARITMATIKA CITRA

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Perbaiki path ke citra Anda jika perlu
image_path = '/content/rivan.jpg'
original_image = cv2.imread(image_path)

# Fungsi untuk menampilkan citra
def display_images(original_image, brightened):
    plt.figure(figsize=(12, 6))

    # Menampilkan citra asli
    plt.subplot(1, 2, 1)
    # Perbaiki nama fungsi menjadi cv2.cvtColor
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title("Citra Asli")
    plt.axis('off')

    # Menampilkan citra setelah pencerahan
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(brightened, cv2.COLOR_BGR2RGB))
    plt.title("Citra Setelah Pencerahan")
    plt.axis('off')

    plt.show()

# Membaca citra


# Parameter pencerahan
brightness_constant = 50

# Operasi pencerahan
brightened_image = cv2.add(original_image, brightness_constant)

# Menampilkan citra sebelum dan setelah pencerahan
display_images(original_image, brightened_image)
````

  - Setiap piksel pada citra diwakili oleh nilai intensitas [B, G, R].

  - Piksel dengan intensitas rendah (gelap) akan lebih terlihat.
  - Piksel yang mendekati putih mungkin kehilangan detail karena saturasi ke nilai maksimum (255).

# Konvolusi Gambar

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload gambar
from google.colab import files
uploaded = files.upload()

# Membaca gambar
image_path = list(uploaded.keys())[0]  # Nama file dari gambar yang diunggah
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Ubah ke RGB untuk tampilan matplotlib

# Menampilkan gambar asli
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Gambar Asli")
plt.axis("off")

# Membuat kernel rata-rata 3x3
kernel = np.ones((3, 3), np.float32) / 9

# Operasi konvolusi
convolved_image = cv2.filter2D(image, -1, kernel)

# Menampilkan hasil konvolusi
plt.subplot(1, 2, 2)
plt.imshow(convolved_image)
plt.title("Hasil Konvolusi (Rata-Rata 3x3)")
plt.axis("off")
plt.show()
````

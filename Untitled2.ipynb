# Soal 1: Jenis-Jenis Tepi

Pertanyaan: Jelaskan empat jenis tepi yang biasa ditemukan dalam citra digital dan beri contoh masing-masing dalam kehidupan sehari-hari. Mengapa tepi dengan derau memerlukan pendekatan khusus untuk pendeteksian?


1. Tepi Curam (Step Edge)
Tepi curam terjadi ketika terdapat perubahan intensitas yang tajam dan langsung antara dua area pada sebuah citra. Perubahan intensitas ini biasanya bersifat diskontinu, di mana satu sisi memiliki nilai intensitas yang berbeda jauh dari sisi lainnya. Contohnya adalah batas antara objek dan latar belakang pada citra.

2. Tepi Landai (Ramp Edge)
Tepi landai terjadi ketika perubahan intensitas terjadi secara bertahap atau perlahan di area tertentu pada citra. Hal ini biasanya ditemukan pada gambar yang memiliki transisi gradasi warna yang halus, seperti bayangan atau objek yang tidak memiliki tepi yang tajam.

3. Tepi Garis (Line Edge)
Tepi garis terjadi ketika terdapat area sempit dengan perubahan intensitas yang signifikan di tengah-tengah dua daerah dengan intensitas yang relatif sama. Biasanya, ini muncul dalam bentuk garis terang pada latar belakang gelap atau garis gelap pada latar belakang terang.

4. Tepi Atap (Roof Edge)
Tepi atap adalah bentuk tepi yang terjadi ketika perubahan intensitas memiliki puncak (seperti atap). Ini biasanya ditemukan pada gambar yang menunjukkan tekstur atau pola yang mengandung garis-garis dengan bentuk "puncak" dan transisi intensitas berbentuk segitiga.

Tepi derau (noise edge) memerlukan pendekatan khusus dalam pemrosesan citra karena sifat derau dapat mengganggu pendeteksian tepi yang sebenarnya.Seperti  Derau Menyebabkan Tepi Palsu,Kesulitan Memisahkan Derau dari Tepi Asli,Gangguan pada Perhitungan Gradien,Efek pada Presisi Deteksi.

# Soal 2: Pendeteksian Tepi Menggunakan Operator Gradien
Pertanyaan: Operator Sobel dan Prewitt digunakan untuk pendeteksian tepi. Berikan contoh implementasi salah satu operator ini dalam Python menggunakan pustaka seperti NumPy atau OpenCV. Jelaskan langkah-langkahnya secara rinci.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar dalam grayscale
image_path = "/content/pcdd.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Terapkan operator Sobel
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradien horizontal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradien vertikal

# Hitung magnitudo gradien
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Konversi ke uint8 agar dapat ditampilkan
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)

# Plot hasil
plt.figure(figsize=(12, 8))
plt.subplot(1, 3, 1), plt.imshow(sobel_x, cmap='gray'), plt.title("Sobel X")
plt.subplot(1, 3, 2), plt.imshow(sobel_y, cmap='gray'), plt.title("Sobel Y")
plt.subplot(1, 3, 3), plt.imshow(sobel_magnitude, cmap='gray'), plt.title("Sobel Magnitude")
plt.show()
````

Berikut adalah penjelasan langkah-langkah pada implementasi operator Sobel secara rinci dari kode sebelumnya:

### **1. Import Pustaka**
```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```
- **`cv2` (OpenCV):** Digunakan untuk membaca gambar, mengaplikasikan operasi Sobel, dan manipulasi citra.

- **`matplotlib.pyplot` (plt):** Untuk menampilkan hasil citra dalam bentuk grafik.

---

### **2. Membaca Gambar**
```python
image = cv2.imread("example.jpg", cv2.IMREAD_GRAYSCALE)
```
- Membaca file gambar `"example.jpg"` menggunakan OpenCV.
- **`cv2.IMREAD_GRAYSCALE`:** Membaca gambar dalam format grayscale (satu kanal intensitas), yang diperlukan untuk deteksi tepi karena hanya intensitas pixel yang relevan.

---

### **3. Menghitung Gradien dengan Operator Sobel**
```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradien horizontal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradien vertikal
```
- **`cv2.Sobel`:** Fungsi untuk menghitung gradien intensitas menggunakan kernel Sobel.
- Parameter:
  - `image`: Input gambar grayscale.
  - `cv2.CV_64F`: Tipe data output, yaitu float64, yang memastikan hasil gradien negatif tidak terpotong.
  - **`1, 0`:** Menghitung gradien horizontal (\(G_x\)) dengan menetapkan arah x = 1 dan y = 0.
  - **`0, 1`:** Menghitung gradien vertikal (\(G_y\)) dengan arah x = 0 dan y = 1.
  - `ksize=3`: Ukuran kernel Sobel (default 3x3, bisa diubah menjadi 5 atau 7).

- **Hasilnya:**
  - `sobel_x`: Citra gradien horizontal.
  - `sobel_y`: Citra gradien vertikal.

---

### **4. Menghitung Magnitudo Gradien**
```python
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)
```
- **`cv2.magnitude`:** Menghitung magnitudo gradien total (\(G = \sqrt{G_x^2 + G_y^2}\)).
- Parameter:
  - `sobel_x`: Gradien horizontal.
  - `sobel_y`: Gradien vertikal.
- Magnitudo gradien digunakan untuk menggabungkan informasi dari gradien horizontal dan vertikal, menghasilkan intensitas tepi pada semua arah.

---

### **5. Konversi ke Format yang Dapat Ditampilkan**
```python
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)
```
- **`cv2.convertScaleAbs`:** Mengubah hasil gradien (yang berupa float dengan nilai positif dan negatif) ke dalam format absolut (uint8) agar dapat ditampilkan sebagai citra.
- Hasil gradien negatif dikonversi menjadi nilai positif untuk keperluan visualisasi.

---

### **6. Menampilkan Hasil**
```python
plt.figure(figsize=(12, 8))
plt.subplot(1, 3, 1), plt.imshow(sobel_x, cmap='gray'), plt.title("Sobel X")
plt.subplot(1, 3, 2), plt.imshow(sobel_y, cmap='gray'), plt.title("Sobel Y")
plt.subplot(1, 3, 3), plt.imshow(sobel_magnitude, cmap='gray'), plt.title("Sobel Magnitude")

```
- **`plt.figure`:** Membuat area untuk visualisasi dengan ukuran tertentu.
- **`plt.subplot`:** Membagi area visualisasi menjadi beberapa bagian (3 dalam hal ini: untuk gradien horizontal, vertikal, dan magnitudo).
- **`plt.imshow`:** Menampilkan gambar dengan skema warna grayscale (`cmap='gray'`).

- **`plt.show`:** Menampilkan hasil visualisasi.

# Soal 3: Perbandingan Operator Pendeteksi Tepi
Pertanyaan: Berikan perbandingan antara operator Sobel, Prewitt, dan Canny berdasarkan akurasi dan kompleksitas komputasi. Dalam situasi apa operator Canny lebih direkomendasikan?

Berikut adalah perbandingan antara **operator Sobel**, **Prewitt**, dan **Canny** berdasarkan **akurasi** dan **kompleksitas komputasi**:

---

### **1. Operator Sobel**
- **Akurasi:**
  - Sobel memberikan hasil yang lebih baik dibandingkan Prewitt karena kernel Sobel memiliki bobot lebih besar di tengah (memberikan prioritas pada piksel pusat).
  - Sensitif terhadap derau karena tidak memiliki mekanisme khusus untuk menangani derau.
- **Kompleksitas Komputasi:**
  - Rendah hingga sedang (hanya menggunakan operasi konvolusi sederhana).
- **Keunggulan:**
  - Cocok untuk deteksi tepi sederhana pada gambar dengan derau rendah.
- **Kelemahan:**
  - Tidak memiliki threshold otomatis untuk menentukan tepi yang signifikan.
  - Kurang presisi dibandingkan Canny dalam mendeteksi tepi halus.

---

### **2. Operator Prewitt**
- **Akurasi:**
  - Mirip Sobel, tetapi lebih sederhana karena bobot kernel Prewitt seragam (tidak memberi prioritas pada piksel pusat).
  - Biasanya memiliki performa yang lebih rendah dibandingkan Sobel dalam hal akurasi, terutama untuk tepi yang tajam atau kompleks.
- **Kompleksitas Komputasi:**
  - Rendah (lebih sederhana dari Sobel karena kernelnya tetap).
- **Keunggulan:**
  - Mudah diimplementasikan dan digunakan untuk deteksi tepi kasar.
- **Kelemahan:**
  - Kurang sensitif terhadap perubahan intensitas halus, sehingga tepi yang lemah mungkin terlewat.

---

### **3. Operator Canny**
- **Akurasi:**
  - Sangat akurat karena melibatkan langkah-langkah berikut:
    1. Penyaringan Gaussian untuk menghaluskan derau.
    2. Perhitungan gradien dengan metode seperti Sobel.
    3. Non-maximum suppression untuk mempertahankan hanya tepi yang signifikan.
    4. Hysteresis thresholding untuk mendeteksi dan menghubungkan tepi secara adaptif.
  - Dapat menangkap tepi yang halus dan kompleks dengan presisi tinggi.
- **Kompleksitas Komputasi:**
  - Tinggi karena melibatkan beberapa langkah tambahan (filter Gaussian, non-maximum suppression, thresholding).
- **Keunggulan:**
  - Mampu mengurangi derau sebelum deteksi.
  - Dapat menangani berbagai jenis citra, termasuk citra dengan derau tinggi.
  - Memberikan hasil yang lebih bersih dan terhubung dengan baik.
- **Kelemahan:**
  - Lebih lambat dibandingkan Sobel dan Prewitt karena proses yang lebih kompleks.
  - Membutuhkan penyesuaian parameter (low dan high threshold) untuk performa optimal.

---

### **Perbandingan Akurasi dan Kompleksitas**

| **Kriteria**       | **Sobel**           | **Prewitt**         | **Canny**           |
|---------------------|---------------------|---------------------|---------------------|
| **Akurasi**         | Sedang              | Rendah-Sedang       | Tinggi              |
| **Ketahanan terhadap Derau** | Rendah              | Rendah              | Tinggi              |
| **Kompleksitas**    | Sedang              | Rendah              | Tinggi              |
| **Konektivitas Tepi** | Kurang Baik         | Kurang Baik         | Sangat Baik         |

---

### **Situasi di Mana Operator Canny Lebih Direkomendasikan**
Operator Canny lebih direkomendasikan dalam situasi berikut:
1. **Ketika Ketahanan terhadap Derau Dibutuhkan:**
   - Jika citra memiliki banyak derau, Canny menggunakan Gaussian blur untuk menghaluskan citra sebelum deteksi, sehingga lebih tahan terhadap gangguan.

2. **Ketika Presisi Tinggi Diperlukan:**
   - Untuk aplikasi seperti segmentasi objek, pengenalan wajah, atau deteksi kontur objek kecil, Canny lebih akurat karena menggunakan non-maximum suppression.

3. **Untuk Citra Kompleks:**
   - Jika citra memiliki banyak detail atau tepi yang halus, Canny dapat mendeteksi tepi secara efektif dan menghubungkan tepi yang terputus.

4. **Aplikasi yang Membutuhkan Konektivitas Tepi:**
   - Pada tugas seperti ekstraksi fitur dalam visi komputer (misalnya, OCR atau deteksi objek), Canny memberikan hasil tepi yang lebih terstruktur dibanding Sobel atau Prewitt.

---

### **Kesimpulan**
- Gunakan **Sobel** atau **Prewitt** untuk tugas sederhana dengan kebutuhan komputasi rendah, terutama jika citra tidak terlalu berderau.
- Pilih **Canny** untuk hasil deteksi tepi yang lebih bersih, akurat, dan terhubung, terutama pada citra dengan derau atau untuk aplikasi yang membutuhkan presisi tinggi.

# Soal 4: Pendeteksian Tepi pada Citra Noisy
Pertanyaan: Sebuah citra memiliki derau "salt-and-pepper". Implementasikan kode Python untuk mendeteksi tepi menggunakan metode Canny dan Sobel setelah melakukan image smoothing menggunakan Gaussian filter. Bandingkan hasilnya dengan citra asli.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menambahkan derau "salt-and-pepper" ke citra
def add_salt_and_pepper_noise(image, amount=0.02):
    noisy_image = image.copy()
    num_salt = int(amount * image.size * 0.5)
    num_pepper = int(amount * image.size * 0.5)

    # Tambahkan "salt" (piksel putih)
    coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 255

    # Tambahkan "pepper" (piksel hitam)
    coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 0

    return noisy_image

# 1. Baca gambar asli dalam grayscale
image_path = "/content/pcdd.jpg"
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 2. Tambahkan derau "salt-and-pepper"
noisy_image = add_salt_and_pepper_noise(original_image, amount=0.05)

# 3. Terapkan Gaussian filter untuk smoothing
smoothed_image = cv2.GaussianBlur(noisy_image, (5, 5), 1.5)

# 4. Deteksi tepi menggunakan Sobel
sobel_x = cv2.Sobel(smoothed_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(smoothed_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)

# 5. Deteksi tepi menggunakan Canny
canny_edges = cv2.Canny(smoothed_image, 100, 200)

# 6. Plot hasil
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1), plt.imshow(original_image, cmap='gray'), plt.title("Original Image")
plt.subplot(2, 3, 2), plt.imshow(noisy_image, cmap='gray'), plt.title("Noisy Image (Salt-and-Pepper)")
plt.subplot(2, 3, 3), plt.imshow(smoothed_image, cmap='gray'), plt.title("Smoothed Image (Gaussian Filter)")
plt.subplot(2, 3, 4), plt.imshow(sobel_magnitude, cmap='gray'), plt.title("Sobel Edge Detection")
plt.subplot(2, 3, 5), plt.imshow(canny_edges, cmap='gray'), plt.title("Canny Edge Detection")
plt.tight_layout()
plt.show()
````

Kurang efektif pada tepi halus atau tepi yang terputus karena ketergantungan pada gradien lokal.

**Canny:**

Memberikan hasil yang lebih bersih, terhubung, dan tajam karena menggunakan langkah-langkah seperti non-maximum suppression dan threshold adaptif.
